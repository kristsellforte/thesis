```python
from fbprophet import Prophet
from fbprophet.diagnostics import cross_validation, performance_metrics

import json
import os
import pickle
```

```python
def get_scores(scores_path):
    try:
        with open(scores_path, 'r') as f:
            return json.load(f)
    except FileNotFoundError:
        return { '0': {} }
```

```python
def save_scores(scores, scores_path):
    try:
        with open(scores_path, 'w') as f:
#             json.load(f)
            json_str = json.dumps(scores)
            f.write(json_str)
    except FileNotFoundError:
        return False
```

```python
def get_version_number(scores):
    return str(int(max(scores.keys())))
```

```python
def get_model(scores):
    version_number = get_version_number(scores)
#     sco
    model_directory = './models/prophet/' + version_number + '/'
#     print(os.listdir(model_path))
    for file_name in os.listdir(model_directory):
        model_path = model_directory + file_name
        model = pickle.load(open(model_path, 'rb'))
        print(model)

#     return model
```

```python
scores_path = './scores/prophet.json'
data_path = './data/train.csv'
scores = get_scores(scores_path)
version_number = get_version_number(scores)
model_directory = './models/prophet/' + version_number + '/'
for file_name in os.listdir(model_directory):
    model_id = file_name.split(".")[0]
    model_path = model_directory + file_name
    model = pickle.load(open(model_path, 'rb'))
    df_cv = cross_validation(model, initial='300 days', period='30 days', horizon='60 days')
    scores[version_number][model_id] = df_cv.to_json(orient = 'records')
save_scores(scores, scores_path)
```
